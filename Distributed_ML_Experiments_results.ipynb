```
class Stats:
  stat_names = ["losses", "grad_norms", "error_norms", "times", "communications",
                "train_losses", "train_grad_norms", "train_accuracies",
                "test_losses", "test_grad_norms", "test_accuracies",
                "smallest_eigens", "largest_eigens",
                "epoch_iters"]
                #"error_norms", "compressions", 

  def __init__(self, name, stats_map):
    self.name = name
    assert len(stats_map) == len(Stats.stat_names)
    for stat in Stats.stat_names:
      self.__dict__[stat] = stats_map[stat]
  
  @staticmethod
  def new(name):
    return Stats(name, {stat:[] for stat in Stats.stat_names})

  @staticmethod
  def dir_name(name):
    return os.path.join(data_dir, "distributed_" + name)

  def git_dir(self):
    return "distributed_" + self.name

  def save(self):
    dir = Stats.dir_name(self.name)
    git_dir = self.git_dir()
    os.makedirs(dir, exist_ok=True) 
    for stat in Stats.stat_names:
      # if not isinstance(self.__dict__[stat], np.ndarray):
      #   assert isinstance(self.__dict__[stat], list)
      #   self.__dict__[stat] = np.array(self.__dict__[stat], dtype=np.float32)
      np.save(os.path.join(dir, stat), self.__dict__[stat])
      git_file_name = os.path.join(git_dir, stat + ".npy")
      git(f"add '{git_file_name}'")
    git(f"commit -m 'Save {self.name}' -q")
    git("pull -q")
    git("commit -m 'Merge' -q")
    git("push -q")

  @staticmethod
  def load(name):
    dir = Stats.dir_name(name)
    if not os.path.exists(dir):
      git("pull -q")
    stats_map = {}
    for stat in Stats.stat_names:
      file_name = os.path.join(dir, stat)
      stats_map[stat] = np.load(file_name + ".npy")
    return Stats(name, stats_map)
```

```
class Dataset(ABC):
  def _split_into_machines(self, ds, mixing_rate, n_machines, num_classes):
    class2indices = [[] for _ in range(num_classes)]
    for i, y in enumerate(ds.targets):
      cls = y if random.random() > mixing_rate else random.randrange(0, num_classes)
      class2indices[cls].append(i)
    
    all_indices = list(itertools.chain(*class2indices))
    chunk_size = (len(ds) + n_machines - 1) // n_machines
    machine_to_indices = to_chunks(all_indices, chunk_size)

    datasets = [Subset(ds, indices) for indices in machine_to_indices]
    return [DataLoader(ds, batch_size=Constants.batch_size_train, shuffle=True, num_workers=Constants.num_load_workers) for ds in datasets]

  def __init__(self, training_ds, test_ds, mixing_rate, n_machines, num_classes):
    self._train_loaders = self._split_into_machines(training_ds, mixing_rate, n_machines, num_classes)
    assert len(self._train_loaders) == n_machines
    self._full_train_loader = DataLoader(training_ds, batch_size=Constants.batch_size_test, shuffle=True, num_workers=Constants.num_load_workers)
    self._test_loader = DataLoader(test_ds, batch_size=Constants.batch_size_test, shuffle=False, num_workers=Constants.num_load_workers)

  @property
  def train_loaders(self) -> List[DataLoader]:
    return self._train_loaders

  @property
  def full_train_loader(self) -> DataLoader:
    return self._full_train_loader

  @property
  def test_loader(self) -> DataLoader:
    return self._test_loader

  @property
  @abstractmethod
  def name(self) -> str:
    pass
```

```
class ModelFactory(ABC):
  @property
  @abstractmethod
  def name(self) -> str:
    pass
    
  @abstractmethod
  def net(self):
    pass
```

```
class DenseModel(ModelFactory):
  def __init__(self, layers):
    self.layers = list(layers)

  @property
  def name(self) -> str:
    return "dense(" + ",".join(map(str, self.layers)) + ")"
    
  def net(self):
    nnLayers = [nn.Flatten()]
    for i in range(len(layers) - 2):
      nnLayers.append(nn.Linear(layers[i], layers[i+1]))
      nnLayers.append(nn.ReLU())
    nnLayers.append(nn.Linear(layers[-2], layers[-1]))
    return nn.Sequential(*nnLayers).to(device)
```

```
def disable_batch_norm(model):
  def replace_bn(module):
    children = [name for name, _ in module.named_children()]
    for child in children:
      target_attr = getattr(module, child)
      if isinstance(target_attr, (nn.BatchNorm1d, nn.BatchNorm2d, nn.BatchNorm3d, nn.LayerNorm)):
        setattr(module, child, nn.Identity())

    for name, child in module.named_children():
      replace_bn(child)

  replace_bn(model)
  return model
```

```
class ResNetBase(ModelFactory, ABC):
  class BasicBlock(nn.Module):
      expansion = 1

      def __init__(self, in_planes, planes, stride=1):
          super().__init__()
          self.conv1 = nn.Conv2d(
              in_planes, planes, kernel_size=3, stride=stride, padding=1, bias=False)
          self.bn1 = nn.BatchNorm2d(planes)
          self.conv2 = nn.Conv2d(planes, planes, kernel_size=3,
                                stride=1, padding=1, bias=False)
          self.bn2 = nn.BatchNorm2d(planes)

          self.shortcut = nn.Sequential()
          if stride != 1 or in_planes != self.expansion*planes:
              self.shortcut = nn.Sequential(
                  nn.Conv2d(in_planes, self.expansion*planes,
                            kernel_size=1, stride=stride, bias=False),
                  nn.BatchNorm2d(self.expansion*planes)
              )

      def forward(self, x):
          out = F.relu(self.bn1(self.conv1(x)))
          out = self.bn2(self.conv2(out))
          out += self.shortcut(x)
          out = F.relu(out)
          return out


  class Bottleneck(nn.Module):
      expansion = 4

      def __init__(self, in_planes, planes, stride=1):
          super().__init__()
          self.conv1 = nn.Conv2d(in_planes, planes, kernel_size=1, bias=False)
          self.bn1 = nn.BatchNorm2d(planes)
          self.conv2 = nn.Conv2d(planes, planes, kernel_size=3,
                                stride=stride, padding=1, bias=False)
          self.bn2 = nn.BatchNorm2d(planes)
          self.conv3 = nn.Conv2d(planes, self.expansion *
                                planes, kernel_size=1, bias=False)
          self.bn3 = nn.BatchNorm2d(self.expansion*planes)

          self.shortcut = nn.Sequential()
          if stride != 1 or in_planes != self.expansion*planes:
              self.shortcut = nn.Sequential(
                  nn.Conv2d(in_planes, self.expansion*planes,
                            kernel_size=1, stride=stride, bias=False),
                  nn.BatchNorm2d(self.expansion*planes)
              )

      def forward(self, x):
          out = F.relu(self.bn1(self.conv1(x)))
          out = F.relu(self.bn2(self.conv2(out)))
          out = self.bn3(self.conv3(out))
          out += self.shortcut(x)
          out = F.relu(out)
          return out


  class Net(nn.Module):
      def __init__(self, block, num_blocks, num_classes=10):
          super().__init__()
          self.in_planes = 64

          self.conv1 = nn.Conv2d(3, 64, kernel_size=3,
                                stride=1, padding=1, bias=False)
          self.bn1 = nn.BatchNorm2d(64)
          self.layer1 = self._make_layer(block, 64, num_blocks[0], stride=1)
          self.layer2 = self._make_layer(block, 128, num_blocks[1], stride=2)
          self.layer3 = self._make_layer(block, 256, num_blocks[2], stride=2)
          self.layer4 = self._make_layer(block, 512, num_blocks[3], stride=2)
          self.linear = nn.Linear(512*block.expansion, num_classes)

      def _make_layer(self, block, planes, num_blocks, stride):
          strides = [stride] + [1]*(num_blocks-1)
          layers = []
          for stride in strides:
              layers.append(block(self.in_planes, planes, stride))
              self.in_planes = planes * block.expansion
          return nn.Sequential(*layers)

      def forward(self, x):
          out = F.relu(self.bn1(self.conv1(x)))
          out = self.layer1(out)
          out = self.layer2(out)
          out = self.layer3(out)
          out = self.layer4(out)
          out = F.avg_pool2d(out, 4)
          out = out.view(out.size(0), -1)
          out = self.linear(out)
          return out
  
  def __init__(self, block, num_blocks, num_classes=10):
    self.block = block
    self.num_blocks = num_blocks
    self.num_classes = num_classes

  def net(self):
    return ResNetBase.Net(self.block, self.num_blocks, self.num_classes).to(device)
    
# def ResNet18():
#     return ResNet(BasicBlock, [2, 2, 2, 2]).to(device)
# def ResNet34():
#     return ResNet(BasicBlock, [3, 4, 6, 3]).to(device)
# def ResNet50():
#     return ResNet(Bottleneck, [3, 4, 6, 3]).to(device)
# def ResNet101():
#     return ResNet(Bottleneck, [3, 4, 23, 3]).to(device)
# def ResNet152():
#     return ResNet(Bottleneck, [3, 8, 36, 3]).to(device)
```

```
class MnistAutoencoder(ModelFactory):
  @property
  def name(self) -> str:
    return "autoencoder"
    
  def net(self):
    encoder = nn.Sequential(
      nn.Conv2d(1, 16, 3, stride=2, padding=1),
      nn.ReLU(),
      nn.Conv2d(16, 32, 3, stride=2, padding=1),
      nn.ReLU(),
      nn.Conv2d(32, 64, 7)
    )
    decoder = nn.Sequential(
      nn.ConvTranspose2d(64, 32, 7),
      nn.ReLU(),
      nn.ConvTranspose2d(32, 16, 3, stride=2, padding=1, output_padding=1),
      nn.ReLU(),
      nn.ConvTranspose2d(16, 1, 3, stride=2, padding=1, output_padding=1),
      nn.Sigmoid()
    )
    return nn.Sequential(encoder, decoder).to(device)
```

```
class CifarAutoencoder(ModelFactory):
  @property
  def name(self) -> str:
    return "autoencoder"
    
  def net(self):
    encoder = nn.Sequential(
      nn.Conv2d(3, 12, 4, stride=2, padding=1),            # [batch, 12, 16, 16]
      nn.ReLU(),
      nn.Conv2d(12, 24, 4, stride=2, padding=1),           # [batch, 24, 8, 8]
      nn.ReLU(),
      nn.Conv2d(24, 48, 4, stride=2, padding=1),           # [batch, 48, 4, 4]
      nn.ReLU(),
    )
    decoder = nn.Sequential(
			nn.ConvTranspose2d(48, 24, 4, stride=2, padding=1),  # [batch, 24, 8, 8]
      nn.ReLU(),
			nn.ConvTranspose2d(24, 12, 4, stride=2, padding=1),  # [batch, 12, 16, 16]
      nn.ReLU(),
      nn.ConvTranspose2d(12, 3, 4, stride=2, padding=1),   # [batch, 3, 32, 32]
      nn.Sigmoid(),
    )
    return nn.Sequential(encoder, decoder).to(device)
```

# Vector Helpers

```
def get_params(model):
  return [param for param in model.parameters() if param.requires_grad and torch.numel(param.data) > 0]
```

```
def n_parameters(model):
  return sum(torch.numel(p) for p in get_params(model))
```

```
def L2(x):
  return torch.linalg.norm(x).item()
```

```
def full_L2(xs):
  assert isinstance(xs, (list, types.GeneratorType)), f"Unexpected type {type(xs)}"
  norms_sq = (L2(x) ** 2 for x in xs)
  return np.sqrt(sum(norms_sq))
```

```
def params_grad_L2(params):
  return full_L2(param.grad for param in params)
```

```
def model_grad_L2(model):
  return params_grad_L2(get_params(model))
```

```
def take_n(seq, n):
  return itertools.islice(seq, int(math.ceil(n)))
```

```
def lst_frac(lst, frac):
  return take_n(lst, frac * len(lst))
```

```
def to_chunks(it, size):
  size = int(math.ceil(size))
  it = iter(it)
  return iter(lambda: tuple(itertools.islice(it, size)), ())
```

```
def map_params(model, f):
  return [f(param) for param in get_params(model)]

def zeros_like_params(model):
  return [torch.zeros_like(param, device=device) for param in get_params(model)]

def randn_like_params(model, mult):
  return [torch.randn_like(param, device=device) + mult for param in get_params(model)]
```

```
def sum_vectors(p1, p2):
  return [x + y for x, y in zip(p1, p2)]

def subtract_vectors(p1, p2):
  return [x - y for x, y in zip(p1, p2)]

def mul_vectors(p1, p2):
  return [x * y for x, y in zip(p1, p2)]
```

```
def model_grads(model):
  return map_params(model, lambda p: p.grad)
```

# Compressors

```
def distribute_results(models, vec, lr):
  with torch.no_grad():
    for model in models:
      for p, v in zip(get_params(model), vec):
        p.data -= v * lr
```

```
class CompressorFactory(ABC):
  @property
  @abstractmethod
  def name(self) -> str:
    pass

  @abstractmethod
  def compressor(self, models) -> CompressorFactory.Compressor:
    pass

  class Compressor(ABC):
    @abstractmethod
    def iter_init(self, models):
      pass
    
    @abstractmethod
    def map(self, model, ind):
      pass

    @abstractmethod
    def reduce(self, sketch1, sketch2):
      pass
    
    @abstractmethod
    def postprocess(self, models, sketch, lr):
      pass

    @property
    def errors(self):
      return self._errors
  
    @property
    def total_communication(self):
      return self.communication

    def apply_error(self, model, ind):
      if self.errors[ind] is None:
        return
      for p, e in zip(get_params(model), self.errors[ind]):
        p.grad += e
      self.errors[ind] = None
```

```
class NoCompressorFactory(CompressorFactory):
  @property
  def name(self) -> str:
    return "regular"

  def compressor(self, models) -> CompressorFactory.Compressor:
    return NoCompressorFactory.NoCompressor(models)

  class NoCompressor(CompressorFactory.Compressor):
    def __init__(self, models):
      n = len(models)
      self._errors = [None] * n
      self.comm_per_iter = n_parameters(models[0]) * n * 2
      self.communication = 0

    def iter_init(self, models):
      pass
    
    def map(self, model, ind):
      return model_grads(model)

    def reduce(self, sketch1, sketch2):
      return sum_vectors(sketch1, sketch2)
    
    def postprocess(self, models, sketch, lr):
      self.communication += self.comm_per_iter
      distribute_results(models, sketch, lr)
```

```
class RandomCompressorFactory(CompressorFactory):
  def __init__(self, frac):
    self.frac = frac

  @property
  def name(self) -> str:
    return f"randomk({self.frac})"

  def compressor(self, model) -> CompressorFactory.Compressor:
    return RandomCompressorFactory.RandomCompressor(model, self.frac)

  class RandomCompressor(CompressorFactory.Compressor):
    def __init__(self, models, frac):
      n = len(models)
      self._errors = [None] * n
      self.frac = frac
      self.params = get_params(models[0])
      self.communication = 0

    def iter_init(self, models):
      def to_bernoulli(p):
        return (torch.rand_like(p, device=device) < self.frac).int()
      self.mask = map_params(models[0], to_bernoulli)
    
    def map(self, model, ind):
      grads = model_grads(model)
      res = mul_vectors(grads, self.mask)
      self._errors[ind] = subtract_vectors(grads, res)
      return res

    def reduce(self, sketch1, sketch2):
      return sum_vectors(sketch1, sketch2)
    
    def postprocess(self, models, sketch, lr):
      self.communication += sum(torch.sum(m).item() for m in self.mask) * len(models) * 2
      distribute_results(models, sketch, lr)
```

# Evaluation

```
class Evaluator(ABC):
  @abstractmethod
  def loss_outputs_len(self, model, batch) -> Tuple[torch.Tensor, torch.Tensor, int]:
    """Returns loss, model outputs and length of the batch"""
    pass
  
  @abstractmethod
  def accuracy(self, outputs, batch) -> float:
    pass
```

```
class ClassifierEvaluator(Evaluator):
  _criterion = nn.CrossEntropyLoss()

  def loss_outputs_len(self, model, batch):
    inputs, labels = batch
    outputs = model(inputs.to(device))
    return ClassifierEvaluator._criterion(outputs, labels.to(device)), outputs, len(labels)

  def accuracy(self, outputs, batch):
    labels = batch[1]
    predicted = torch.max(outputs, 1)[1]
    return 100 * (predicted.to(cpu) == labels).sum().item() / len(labels)
```

```
class BertEvaluator(Evaluator):
  def loss_outputs_len(self, model, batch):
    inputs, labels, bert_attention_marks = batch
    outputs = model(inputs.to(device),
                    labels=labels.to(device),
                    attention_mask=bert_attention_marks.to(device),
                    token_type_ids=None)
    return outputs[0], outputs[1], len(labels)

  def accuracy(self, outputs, batch):
    labels = batch[1]
    predicted = torch.max(outputs, 1)[1]
    return 100 * (predicted.to(cpu) == labels).sum().item() / len(labels)
```

```
class AutoencoderEvaluator(Evaluator):
  _criterion = nn.MSELoss()

  def loss_outputs_len(self, model, batch):
    inputs = batch[0].to(device)
    labels = batch[1]
    outputs = model(inputs)
    return AutoencoderEvaluator._criterion(outputs, inputs), outputs, len(labels)

  def accuracy(self, outputs, batch):
    return 0
```

# Args

```
def copy_with(obj, **kwargs):
  args = dict(obj.__dict__)
  for k,v in kwargs.items():
    args[k] = v
  return obj.__class__(**args)

def isnumber(x):
  return isinstance(x, (int, float))
```

```
def none_or_instance(obj, cls):
  return obj is None or isinstance(obj, cls)

class OptimizerArgs:
  def __init__(self, *, learning_rate, epochs, noise, weight_decay):
    self.learning_rate = learning_rate
    self.epochs = epochs
    self.noise = noise
    self.weight_decay = weight_decay
  
  def copy(self, **kwargs):
    return copy_with(self, **kwargs)

  def check(self):
    assert isnumber(self.learning_rate)
    assert isinstance(self.epochs, int)
    assert isnumber(self.noise)
    assert isnumber(self.weight_decay)

  def __repr__(self):
    return f"opt(lr={self.learning_rate},epochs={self.epochs},noise={self.noise},wd={self.weight_decay})"


class Args:
  def __init__(self, *, dataset, model_f, batch_normalization, optimizer_args, evaluator, compressor_f, reset_error, eigen_epochs, n_runs):
    assert none_or_instance(dataset, Dataset)
    self.dataset = dataset

    assert none_or_instance(model_f, ModelFactory)
    self.model_f = model_f

    assert none_or_instance(batch_normalization, bool)
    self.batch_normalization = batch_normalization

    assert none_or_instance(optimizer_args, OptimizerArgs)
    self.optimizer_args = optimizer_args

    assert none_or_instance(evaluator, Evaluator)
    self.evaluator = evaluator

    assert none_or_instance(compressor_f, CompressorFactory)
    self.compressor_f = compressor_f

    assert none_or_instance(reset_error, bool)
    self.reset_error = reset_error

    assert none_or_instance(eigen_epochs, int)
    self.eigen_epochs = eigen_epochs

    assert none_or_instance(n_runs, int)
    self.n_runs = n_runs
  
  def copy(self, **kwargs):
    return copy_with(self, **kwargs)

  def check(self):
    assert not self.batch_normalization # Can't handle BN in distributed settings

    assert isinstance(self.dataset, Dataset)
    assert isinstance(self.model_f, ModelFactory)
    assert isinstance(self.batch_normalization, bool)
    assert isinstance(self.optimizer_args, OptimizerArgs)
    assert isinstance(self.evaluator, Evaluator)
    assert isinstance(self.compressor_f, CompressorFactory)
    assert isinstance(self.reset_error, bool)
    assert isinstance(self.eigen_epochs, int) and self.eigen_epochs >= 0
    assert isinstance(self.n_runs, int) and self.n_runs >= 1
    self.optimizer_args.check()
  
  def __repr__(self):
    bn_str = "BN" if self.batch_normalization else "noBN"
    return f"ds={self.dataset.name}_model={self.model_f.name}_{bn_str}_opt={self.optimizer_args}_compr={self.compressor_f.name}_eigens={self.eigen_epochs}_nruns={self.n_runs}"
```

## Information after epoch

```
def cur_stats(model, data_loader, evaluator, frac) -> Tuple[float, float, float]:
  """Returns loss, gradient norm and accuracy computed using a fraction of the dataset"""
  model.zero_grad()
  total_loss = 0
  total_len = 0
  total_acc = 0
  for batch in lst_frac(data_loader, frac):
    loss, outputs, cur_len = evaluator.loss_outputs_len(model, batch)
    total_acc += evaluator.accuracy(outputs, batch) * cur_len
    total_len += cur_len
    loss *= cur_len
    loss.backward()
    total_loss += loss.item()

  norm = model_grad_L2(model)
  return total_loss / total_len, norm / total_len, total_acc / total_len
```

```
def epoch_summary(model, dataset, evaluator, power_method, frac, stats):
  """Adds epoch information to stats and prints it"""
  iter = len(stats.losses[-1])
  stats.epoch_iters[-1] += [iter]

  train_loss, train_grad_norm, train_acc = cur_stats(model, dataset.full_train_loader, evaluator, frac)
  stats.train_losses[-1] += [train_loss]
  stats.train_grad_norms[-1] += [train_grad_norm]
  stats.train_accuracies[-1] += [train_acc]

  test_loss, test_grad_norm, test_acc = cur_stats(model, dataset.test_loader, evaluator, 1)
  stats.test_losses[-1] += [test_loss]
  stats.test_grad_norms[-1] += [test_grad_norm]
  stats.test_accuracies[-1] += [test_acc]

  if power_method is not None:
    e1, e2 = smallest_and_largest_eigenvalues(power_method, dataset.full_train_loader)
    stats.smallest_eigens[-1] += [e1]
    stats.largest_eigens[-1] += [e2]
    eigen_str = f" e1: {e1} e2: {e2}"
  else:
    eigen_str = ""

  print(f"TrainLoss: {train_loss:.4f} TrainGradNorm: {train_grad_norm:.4f} TrainAcc: {train_acc:.2f} TestLoss: {test_loss:.4f} TestGradNorm: {test_grad_norm:.4f} TestAcc: {test_acc:.2f}" + eigen_str, flush=True)
```

## Training Functions

```
def distributed_step(models, train_loaders, evaluator, noise, compressor, lr):
  final_sketch = None
  n = len(models)
  full_grad = zeros_like_params(models[0])
  full_loss = 0
  compressor.iter_init(models)
  for i in range(n):
    model = models[i]
    batch = next(train_loaders[i])

    model.zero_grad()
    loss = evaluator.loss_outputs_len(model, batch)[0]
    full_loss += loss.item()
    loss.backward()

    with torch.no_grad():
      full_grad = sum_vectors(full_grad, model_grads(model))
      for param in get_params(model):
        param.grad += torch.randn_like(param.grad) * noise

      compressor.apply_error(model, i)
      local_sketch = compressor.map(model, i)
      final_sketch = local_sketch if final_sketch is None else compressor.reduce(final_sketch, local_sketch)

  with torch.no_grad():
    for x in final_sketch:
      x /= n
    compressor.postprocess(models, final_sketch, lr)

    full_error = zeros_like_params(models[0])
    for err in compressor.errors:
      if err is not None:
        full_error = sum_vectors(full_error, err)

  return full_loss / n, full_L2(full_grad) / n, full_L2(full_error) / n
```

```
def train(args: Args, stats: Stats, run: int):
  for stat in Stats.stat_names:
    stats.__dict__[stat].append([])

  assert not args.batch_normalization
  models = [disable_batch_norm(args.model_f.net()) for _ in args.dataset.train_loaders]

  with torch.no_grad():
    for model in models:
      for p1, p2 in zip(get_params(model), get_params(models[0])):
        p1.data[:] = p2.data
      model.train()

  compressor = args.compressor_f.compressor(models)
  num_parameters = n_parameters(models[0])
  print("Number of parameters:", num_parameters)

  optimizer_args = args.optimizer_args

  if args.eigen_epochs > 0:
    power_method = PowerMethod(models[0], args.evaluator, args.eigen_epochs)
  else:
    power_method = None

  start_time = time()
  for epoch in range(optimizer_args.epochs):
    epoch_summary(models[0], args.dataset, args.evaluator, power_method, 0.2, stats)
    full_losses, full_grad_norms, full_error_norms = 0, 0, 0
    epoch_train_loaders = [iter(loader) for loader in args.dataset.train_loaders]
    min_len = min(len(loader) for loader in args.dataset.train_loaders)
    progress_bar = tqdm(range(min_len), position=0, leave=True, miniters=20)
    for i in progress_bar:
      loss, grad_norm, error_norm = distributed_step(models, epoch_train_loaders, args.evaluator, optimizer_args.noise, compressor, optimizer_args.learning_rate)
      stats.losses[-1] += [loss]
      stats.grad_norms[-1] += [grad_norm]
      stats.error_norms[-1] += [error_norm]
      stats.communications[-1] += [compressor.total_communication]
      stats.times[-1] += [time() - start_time]

      full_losses += loss
      full_grad_norms += grad_norm
      full_error_norms += error_norm

      progress_bar.set_description(f"{run} {epoch:2} {i:3} Loss: {full_losses / (i + 1):.4f} GradNorm: {full_grad_norms / (i + 1):.4f} " +
                                   f"ErrNorm: {full_error_norms / (i + 1):.4f} Comm: {compressor.total_communication:,} Time: {stats.times[-1][-1]:.2f}",
                                   refresh=False)

    if args.reset_error:
      with torch.no_grad():
        full_error = zeros_like_params(models[0])
        for err in compressor.errors:
          if err is not None:
            full_error = sum_vectors(full_error, err)
        for x in full_error:
          x /= len(models)
        distribute_results(models, full_error, optimizer_args.learning_rate)
        compressor.error = [None] * len(compressor.error)

  epoch_summary(models[0], args.dataset, args.evaluator, power_method, 1, stats)
```

```
def averaged(data, period):
  res = []
  sum = 0
  for i in range(data.shape[1]):
    sum += data[:, i]
    if i >= period:
      sum -= data[:, i - period]
      res.append(sum / period)
    else:
      res.append(sum / (i+1))
  res = np.transpose(np.array(res))
  return res
```

```
def compare_stat(plot_name, stats, names):
  os.makedirs("plots", exist_ok=True)
  line_colors = ["red", "green", "blue", "purple"]
  line_styles = ["-", "--", "-.", ":"]
  figsize = (5, 3)
  fontsize = 18
  epoch_xs = np.arange(len(stats[0].train_losses[0]))
  xs = np.arange(len(stats[0].losses[0])) / (stats[0].epoch_iters[0, 1] - stats[0].epoch_iters[0, 0])

  fig = plt.figure(figsize=figsize)
  ax = fig.add_subplot(1, 1, 1)
  plt.yscale("log")
  ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
  ax.yaxis.set_minor_formatter(ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
  for i in range(len(stats)):
    # print(len(epoch_xs), stats[i].test_losses.shape)
    plot_with_error(epoch_xs, stats[i].test_losses, color=line_colors[i], label=names[i], linestyle=line_styles[i], linewidth=4)
    plt.xlabel("Epoch", fontsize=fontsize)
    plt.ylabel("Test Loss (log-sc.)", fontsize=fontsize)
  plt.savefig(f"plots/{plot_name}_test_loss.pdf", bbox_inches='tight')

  fig = plt.figure(figsize=figsize)
  ax = fig.add_subplot(1, 1, 1)
  plt.yscale("log")
  ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
  ax.yaxis.set_minor_formatter(ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
  for i in range(len(stats)):
    # print(len(epoch_xs), stats[i].test_losses.shape)
    plot_with_error(epoch_xs, 100 - stats[i].test_accuracies, color=line_colors[i], label=names[i], linestyle=line_styles[i], linewidth=4)
    plt.xlabel("Epoch", fontsize=fontsize)
    plt.ylabel("Test Error, % (log-sc.)", fontsize=fontsize)
  plt.savefig(f"plots/{plot_name}_test_error.pdf", bbox_inches='tight')


  fig = plt.figure(figsize=figsize)
  ax = fig.add_subplot(1, 1, 1)
  plt.yscale("log")
  ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: '{:g}'.format(y)))
  for i in range(len(stats)):
    plot_with_error(epoch_xs, stats[i].train_losses, color=line_colors[i], label=names[i], linestyle=line_styles[i], linewidth=4)
    plt.xlabel("Epoch", fontsize=fontsize)
    plt.ylabel("Train Loss (log-sc.)", fontsize=fontsize)
  plt.savefig(f"plots/{plot_name}_train_loss.pdf", bbox_inches='tight')

  fig = plt.figure(figsize=figsize)
  for i in range(len(stats)):
    plot_with_error(epoch_xs, stats[i].train_grad_norms, color=line_colors[i], label=names[i], linestyle=line_styles[i], linewidth=4)
    plt.xlabel("Epoch", fontsize=fontsize)
    plt.ylabel("Grad norm", fontsize=fontsize)
  plt.savefig(f"plots/{plot_name}_grad_norm.pdf", bbox_inches='tight')

  fig = plt.figure(figsize=figsize)
  plt.yscale("log")
  plt.xscale("log")
  for i in range(len(stats)):
    plot_with_error(xs, stats[i].communications, color=line_colors[i], label=names[i], linestyle=line_styles[i], linewidth=4)
    plt.xlabel("Epoch (log-sc.)", fontsize=fontsize)
    plt.ylabel("Communication (log-sc.)", fontsize=fontsize)
  plt.savefig(f"plots/{plot_name}_communication.pdf", bbox_inches='tight')

  fig = plt.figure(figsize=figsize)
  plt.yscale("log")
  # plt.xscale("log")
  for i in range(len(stats)):
    iters = [int(x) for x in stats[i].epoch_iters[0]]
    iters[-1] -= 1
    plot_with_error(stats[i].communications[0][iters], stats[i].train_losses, color=line_colors[i], label=names[i], linestyle=line_styles[i], linewidth=4)
  plt.xlabel("Communication", fontsize=fontsize)
  plt.ylabel("Train Loss", fontsize=fontsize)
  plt.savefig(f"plots/{plot_name}_train_loss_vs_comm.pdf", bbox_inches='tight')


  ax = plt.axes()
  fig = plt.figure(figsize = (18,0.5))
  handles, labels = ax.get_legend_handles_labels()
  legend_prop = {'weight' : 'bold', 'size' : 17}
  fig.legend(handles, labels, loc='center', mode='expand', ncol = 4,
          handlelength = 3, handletextpad = 0.3, prop = legend_prop)
  fig.savefig(f"plots/{plot_name}_legend.pdf")
```

```
def tune_args(args, *, noise=None, compressor_f=None, epochs=None, learning_rate=None, weight_decay=None, reset_error=None, eigen_epochs=None, n_runs=None):
  if weight_decay is not None:
    args = args.copy(optimizer_args=args.optimizer_args.copy(weight_decay=weight_decay))
  if learning_rate is not None:
    args = args.copy(optimizer_args=args.optimizer_args.copy(learning_rate=learning_rate))
  if epochs is not None:
    args = args.copy(optimizer_args=args.optimizer_args.copy(epochs=epochs))
  if noise is not None:
    args = args.copy(optimizer_args=args.optimizer_args.copy(noise=noise))
  if compressor_f is not None:
    args = args.copy(compressor_f=compressor_f)
  if reset_error is not None:
    args = args.copy(reset_error=reset_error)
  if eigen_epochs is not None:
    args = args.copy(eigen_epochs=eigen_epochs)
  if n_runs is not None:
    args = args.copy(n_runs=n_runs)

  return args
```
